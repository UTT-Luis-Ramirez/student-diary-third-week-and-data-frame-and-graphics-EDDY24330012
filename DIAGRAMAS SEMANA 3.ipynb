##QUE ES UN HISTOGRAMA?
Un histograma es un gráfico de barras que permite representar la frecuencia de un valor estadístico dento de un grupo o poblacion de estudio. Es decir, permite observar desde un plano general la distribución de una caracteristica cuantitativa y continua, o compara los resultados de un proceso específico.
Consta de dos ejes. uno vertical (Y) donde se representan las frecuencias, es decir, la cantidad de veces que se repite un evento; y el eje horizontal (X) se representan los rasgos o caracteristicas que se estudian en la muestra.

---



**Lo que caracteriza al histograma es:**

*-Permite cotejar la frecuencia estadistica de un suceso con una variable cuantitativa y continua*

*-ubican siempre en el eje Y la frecuencia estadistica y en el eje X la variable a estudiar*

---



**Tipos de histogramas**

*-Histograma de frecuencias absolutas*

*-Histograma de frecuenciasa relativas*

*-Histograma de intervalos desiguales*

*-Histograma de funcion de densidad*


---


##Explicacion codigo Histograma:
* Se importa la biblioteca **matplotlib** y se le da el alias **plt**, que es un estándar en Python para crear gráficos.*

* Se importa **numpy** **(alias np)**, que se usa para crear datos **numéricos** (en este caso, un conjunto de datos **aleatorios**).*

* **data = np.random.randn(1000):** Crea un **arreglo** de 1000 números **aleatorios** distribuidos de manera normal

**ARREGLO:** *También conocido como ***array*** en inglés, es una estructura de datos que almacena una colección de elementos en una secuencia ordenada. Cada elemento en un arreglo puede ser de cualquier tipo de dato (números, cadenas de texto, objetos, etc.)*

* **data** es el conjunto de datos que se va a graficar. **bins=30** divide el rango de los datos en 30 intervalos **(barras del histograma)**, y **edgecolor='red'** define el color del borde de cada barra.

* **plt.title('Histograma Ejemplo'):** Establece el título del gráfico.

* **plt.xlabel('Valor'):** Establece la etiqueta del eje x, que en este caso es el "Valor".
* **plt.ylabel('Frecuencia'):** Establece la etiqueta del eje y, que indica la "Frecuencia" de los valores en el histograma.
* **plt.show():** Muestra el gráfico.

```python
import matplotlib.pyplot as plt
import numpy as np

# Crear datos de ejemplo
data = np.random.randn(1000)

# Crear el histograma
#
plt.hist(data, bins=30, edgecolor='red')
plt.title('Histograma Ejemplo')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

##QUE ES UN DIAGRAMA DE TALLO Y HOJA?
Un diagrama de tallo y hoja es una herramienta utilizada en estadísticas para visualizar la distribución de un conjunto de datos. Divide cada dato en dos partes: el "tallo" (que representa la parte más significativa) y la "hoja" (que representa la parte menos significativa). Esta técnica permite ver rápidamente la forma en que los datos están distribuidos, incluyendo su rango, centros y dispersión.
Este diagrama se construye separando valores de cada observación en dos partes por medio de una línea vertical. La primera (tallo) se ubica a la izquierda de la línea vertical y generalmente corresponde a la decena del valor que toma la variable; y la segunda (hoja) se ubica a la derecha de la línea vertical y generalmente corresponde a la unidad del valor que toma la variable. Si se tiene muchas hojas en cada tallo, es posible separarlas en dos tallos.

---



*Lo que carteriza a un diagrama de tallo y hojas es:**

*-Muestra claramente la distribución*

*-Divivde los datos en dos partes el tallo (la parte más significativa) y la hoja la parte menos significativa.*

---





#Diagrama de Tallo y Hoja
* import matplotlib.pyplot as plt: Se importa la biblioteca matplotlib para la creación de gráficos.
* data = [12, 23, 45, 56, 67, 89, 90, 120, 130, 143]: Define una lista con 10 valores numéricos de ejemplo para graficar en el diagrama de tallo y hoja.

* plt.stem(data, use_line_collection=True): Crea el diagrama de tallo y hoja. data son los valores numéricos a graficar.

* plt.title('Diagrama de tallo y hoja'): Establece el título del gráfico.

* plt.xlabel('Índice'): Establece la etiqueta del eje x, que indica el "Índice" de los datos en la lista.

* plt.ylabel('Valor'): Establece la etiqueta del eje y, que indica el "Valor" de los datos.

* plt.show(): Muestra el gráfico.

```python
import matplotlib.pyplot as plt

# Crear datos de ejemplo
data = [12, 23, 45, 56, 67, 89, 90, 120, 130, 143]

# Crear diagrama de tallo y hoja
plt.stem(data)
plt.title('Diagrama de tallo y hoja')
plt.xlabel('Índice')
plt.ylabel('Valor')
plt.show()
```

##QUE ES UN DIAGRAMA DE CAJA
Es una representacion grafica que permite visualizar la distribucion de un conjunto de datos, resaltando sus cuartiles, la mediana, y valores atípicos, este tipo de grafico es muy util para entender la dispersion y simetria de los datos a simple vista.
El lado inferior del rectángulo representa el primer cuartil, y el lado superior, el tercer cuartil. En consecuencia, la altura de la caja representa el rango intercuartilico. La línea horizontal a través de la caja es la mediana. Las líneas verticales que sobresalen de la caja, el 'bigotes', se extienden, respectivamente, hasta al mínimo y el massimo del conjunto de datos, siempre que estos valores no difieren de la media de más de una vez y media el rango intercuartílico. Los extremos de los bigotes están marcados por dos líneas horizontales cortas. Los valores, indicados por puntos, respectivamente, por debajo y por encima de los bigotes inferior y superior se consideran valores atípicos.

---



**Lo que carteriza a un diagrama de tallo y hojas es:**

*-Distribucion y resumen estadistico*

*-visualizacion de la dispersión*

*-identificacion de valores atipicos*

*-comparacion ente conjuntos de datos*


---




##Diagrama de caja
* import seaborn as sns: Se importa la biblioteca seaborn (alias sns), que es una extensión de matplotlib y proporciona gráficos estadísticos más atractivos y fáciles de usar.

* import numpy as np: Se importa numpy para generar datos numéricos aleatorios.
data = np.random.randn(100): Crea un conjunto de 100 números aleatorios con distribución normal estándar (media 0, desviación estándar 1).

* sns.boxplot(data=data): Crea el diagrama de caja usando seaborn. El parámetro data=data es el conjunto de datos que se va a graficar.

* plt.title('Diagrama de caja'): Establece el título del gráfico.

* plt.show(): Muestra el gráfico.

```python
import seaborn as sns
import numpy as np

# Crear datos de ejemplo
data = np.random.randn(100)

# Crear el diagrama de caja
sns.boxplot(data=data)
plt.title('Diagrama de caja')
plt.show()
```
